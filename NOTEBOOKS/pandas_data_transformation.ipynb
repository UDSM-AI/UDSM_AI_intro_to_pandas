````python
import pandas as pd


df = pd.read_csv("datasets/Iris.csv")
df.head()
````

# Applying a Function to a Column:

````python
# Applying a function to a column
df['magnified_SepalLengthCm'] = df['SepalLengthCm'].apply(lambda x: x * 2)
````

The code above create a new column called 'new_column' by applying a lambda function to each element in the 'column' of the DataFrame df. The lambda function takes each value of 'column' and multiplies it by 2, assigning the result to the corresponding row in 'new_column'.
