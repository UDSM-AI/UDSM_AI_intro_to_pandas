# Data Selection

load data/ create a dataframe

```python
import pandas as pd

df = pd.read_csv("datasets/titanic.csv")

# df.tail()

df['new_class'] = df['Class']
df.head()
```

```python
# Selecting a single column
column = df['Destination']

column
```

```python
# Selecting multiple columns
columns = df[['Sex', 'Age', 'Destination']]

columns
```

```python

# Selecting rows based on a condition
filtered_df = df[df['Age'] > 10]

filtered_df
```

# Data Sorting

```python
# Sorting by a single column
sorted_df = df.sort_values('Age')

sorted_df
```

```python

# Sorting by multiple columns
sorted_df = df.sort_values(['Age', 'PassengerId'], ascending=[True, True])

sorted_df
```

```python
# Calculating the mean of a column
mean_value = df['Age'].mean()
print("Mean Age: ", mean_value)

# Calculating the sum of a column
sum_value = df['Survived'].sum()
print("Total Survived: ", sum_value)

# Grouping by a column and calculating the mean
grouped_df = df.groupby('Sex')['Age']

grouped_df.mean()

```
