```python
import pandas as pd
```

```python
# read data
dataX = pd.read_csv("./challenge_initial_dataset.csv")
dataX.head()
```

```python
# information about our dataset
dataX.info()
```

```python
dataX.describe()
```

```python
dataX[~dataX['Comorbidities'].isnull()]
```

```python
# I use "No Comobid" to fill null values found in Comobidities column. I do this b'cause patient can have a disease without its Comorbidities
dataX["Comorbidities"] = dataX["Comorbidities"].fillna("No Comobid")
```

```python
dataX[~dataX['Comorbidities'].isnull()]
```

```python
# primary diagnosis is the required filled so if it's null should be removed
dataX = dataX.dropna(subset=["Primary Diagnosis"])

# Fill null value with "Not provided" in Medication History and Readmission Status column.
dataX["Medication History"] = dataX["Medication History"].fillna("Not provided")
dataX["Readmission Status"] = dataX["Readmission Status"].fillna("Not provided")
```

```python
# check if there's any duplicated data
dataX[dataX.duplicated()]
```
